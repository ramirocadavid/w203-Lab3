# Bivariate regression (no transformations)

```r
bivariate = function(x_col, y_col, data){
    x = data[,x_col]
    y = data[,y_col]
    model = lm(y ~ x)
    cat(paste("bivariate regression for", x_col, "and", y_col,"\n"))
    print(model$coefficients)
    cat(paste("r-squared =", summary(model)$r.square))
    plot(x, y, main = paste(x_col, "vs", y_col), xlab = x_col, ylab = y_col)
    abline(model)
    return(model)
}
```

```r
bivariate_prbarr = bivariate("prbarr", "crmrte", crime)
```

```r
bivariate_prbconv = bivariate("prbconv", "crmrte", crime)
```

```r
bivariate_taxpc = bivariate("taxpc", "crmrte", crime)
```

```r
bivariate_urban = bivariate("urban", "crmrte", crime)
```

```r
bivariate_pctmin80 = bivariate("pctmin80", "crmrte", crime)
```

```r
bivariate_wser = bivariate("wser", "crmrte", crime)
```

```r
bivariate_pctymle = bivariate("pctymle", "crmrte", crime)
```

```r
stargazer(bivariate_prbarr,   #1
          bivariate_prbconv,  #2
          bivariate_prbpris,  #3
          bivariate_polpc,    #4
          bivariate_density,  #5
          bivariate_taxpc,    #6 
          bivariate_urban,    #7
          bivariate_pctmin80, #8 
          bivariate_wser,     #9
          bivariate_pctymle,  #10
          type = "text", 
          report = "vc",
          title = "Bivariate Linear Models Predicting Crime Rate",
          keep.stat = c("rsq", "n"),
          omit.table.layout = "n")
```

# Bivariate Regression By Category

```r
crime_urban = crime[crime$urban == 1,]
dim(crime_urban)
```

```r
crime_noturban = crime[crime$urban == 0,]
dim(crime_noturban)
```

```r
crime_west = crime[crime$west == 1,]
dim(crime_west)
```

```r
crime_central = crime[crime$central == 1,]
dim(crime_central)
```

```r
crime_otherloc = crime[crime$west == 0 & crime$central == 0,]
dim(crime_otherloc)
```

### urban

```r
stargazer(bivariate_urban,          #1
          bivariate_noturban,       #2
          bivariate_west,           #3
          bivariate_central,        #4
          bivariate_crime_otherloc, #5
          type = "text", 
          report = "vc",
          title = paste(xcol, "vs crime rate"),
          keep.stat = c("rsq", "n"),
          omit.table.layout = "n")
```
