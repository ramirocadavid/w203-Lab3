### Load the data

```r
crime = read.csv("crime_v2.csv")
```

```r
#notice that prbconv is a Factor, while the rest are int or num
str(crime)
```

```r
#look at the dimensions of the dataframe
#there are 97 rows and 25 columns
dim(crime)
```

```r
head(crime)
```

### Remove the NA rows

```r
#looking at the tail of the dataframe shows that there are many NA rows
tail(crime, 7)
```

```r
#remove the NA rows
crime = crime[0:91,]
```

```r
#check to see that the NA rows were correctly removed
tail(crime)
```

```r
#check the dimensions of the dataframe again
#there are 91 rows and 25 columns
dim(crime)
```

### Remove duplicated row

```r
#check number of unique values
#find that there is 90 unique values for county but there are 91 rows
length(unique(crime$county))
```

```r
#find the duplicated value
crime[duplicated(crime$county),]
```

```r
#check that the duplicated value is actually duplicated
crime[crime$county == 193,]
```

```r
#all 25 values match, confirming that the rows are duplicated
sum(crime[88,] == crime[89,])
```

```r
#remove duplicated row
crime = crime[!duplicated(crime$county),]
```

```r
#check the number of observations for county
length(crime$county)
```

### Convert prbconv from a factor to numeric data

```r
crime[,"prbconv"]
```

```r
#check if it is numeric
is.numeric(crime[,"prbconv"])
```

```r
#convert to numeric
crime[,"prbconv"] = as.numeric(levels(crime[,"prbconv"])[crime[,"prbconv"]])
```

```r
#check to see if conversion to numeric was successful
is.numeric(crime[,"prbconv"])
str(crime)
```

### View variables

```r
#adjusts plot sizes
options(repr.plot.width=4, repr.plot.height=4)
```

```r
#function to get info for all columns
for(colname in colnames(crime)){
    hist(crime[,colname], main=paste("Histogram of",colname), xlab=colname)
    cat(paste(colname, "has", length(na.omit(crime[,colname])), "observations and", sum(is.na(crime[,colname])), "nulls\n"))
    print(summary(crime[,colname]))  
}
```
