# Bivariate regression (no transformations)

```r
bivariate = function(x_col, y_col, data){
    x = data[,x_col]
    y = data[,y_col]
    model = lm(y ~ x)
    cat(paste("bivariate regression for", x_col, "and", y_col,"\n"))
    print(model$coefficients)
    cat(paste("r-squared =", summary(model)$r.square))
    plot(x, y, main = paste(x_col, "vs", y_col), xlab = x_col, ylab = y_col)
    abline(model)
    return(model)
}
```

```r
bivariate_prbarr = bivariate("prbarr", "crmrte", crime)
```

```r
resid_bivariate_prbarr = resid(bivariate_prbarr)
hist(resid_bivariate_prbarr)
summary(resid_bivariate_prbarr)
```

# Bivariate regression (with transformations)

```r
bivariate_xlog = function(x_col, y_col, data){
    x = log(data[,x_col])
    y = data[,y_col]
    model = lm(y ~ x)
    cat(paste("bivariate regression for log", x_col, "and", y_col,"\n"))
    print(model$coefficients)
    cat(paste("r-squared =", summary(model)$r.square))
    plot(x, y, main = paste("log", x_col, "vs", y_col), xlab = paste("log", x_col), ylab = y_col)
    abline(model)
    return(model)
}
```

```r
bivariate_logprbarr = bivariate_xlog("prbarr", "crmrte", crime)
```

```r
resid_bivariate_logprbarr = resid(bivariate_logprbarr)
hist(resid_bivariate_logprbarr)
summary(resid_bivariate_logprbarr)
```

```r
bivariate_logprbconv = bivariate_xlog("prbconv", "crmrte", crime)
```

```r
resid_bivariate_logprbconv = resid(bivariate_logprbconv)
hist(resid_bivariate_logprbconv)
summary(resid_bivariate_logprbconv)
```

```r
bivariate_logprbpris = bivariate_xlog("prbpris", "crmrte", crime)
```

```r
resid_bivariate_logprbpris = resid(bivariate_logprbpris)
hist(resid_bivariate_logprbpris)
summary(resid_bivariate_logprbpris)
```

```r
bivariate_ylog = function(x_col, y_col, data){
    x = data[,x_col]
    y = log(data[,y_col])
    model = lm(y ~ x)
    cat(paste("bivariate regression for", x_col, "and log", y_col,"\n"))
    print(model$coefficients)
    cat(paste("r-squared =", summary(model)$r.square))
    plot(x, y, main = paste(x_col, "vs log", y_col), xlab = x_col, ylab = paste("log", y_col))
    abline(model)
    return(model)
}
```

```r
bivariate_logcrime_prbarr = bivariate_ylog("prbarr", "crmrte", crime)
```

```r
resid_bivariate_logcrime_prbarr = resid(bivariate_logcrime_prbarr)
hist(resid_bivariate_logcrime_prbarr)
summary(resid_bivariate_logcrime_prbarr)
```

# Model 1

### no transformation

```r
model1 = lm(crmrte ~ prbarr + prbconv + prbpris, data = crime)
summary(model1)$coefficient
summary(model1)$r.squared
summary(model1)$adj.r.squared
AIC(model1)
```

```r
hist(resid(model1), breaks = 15)
summary(resid(model1))
0.002389/sd(resid(model1))
sd(resid(model1))
```

```r
plot(model1)
```
